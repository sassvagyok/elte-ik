1.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2.

```python
df = pd.read_csv("spotify_yt.csv")
```

3.

```python
print(df.info())
print(df.describe())
```

4.

```python
print(df.isnull().sum())
```

5.

```python
colums_to_fill = ["Loudness", "Acousticness", "Instrumentalness", "Energy", "Danceability"]

for c in colums_to_fill:
    means = df.groupby("Artist")[c].transform("mean")
    df[c] = df[c].fillna(means)

print(df.isnull().sum())
```

6.

```python
colums_to_check = ["Speechiness", "Liveness", "Valence", "Tempo", "Key", "Duration_ms"]
print(df[df[colums_to_check].isnull().any(axis=1)])

df.dropna(subset=colums_to_check, inplace=True)
print(df.isnull().sum())
```

7.

```python
print(df.dtypes)

columns_to_int = ["Duration_ms", "Key", "Views", "Likes", "Comments", "Stream"]

for c in columns_to_int:
    df[c] = df[c].fillna(-1).astype(int)

columns_to_bool = ["Licensed", "official_video"]
for c in columns_to_bool:
    df[c] = df[c].astype("boolean")
    df[c] = df[c].fillna(False)
```

8.

```python
df = df[df["Stream"] != -1]
df.drop(columns=["Url_spotify", "Uri"], inplace=True)
```

9.

```python
df.to_csv("spotify_yt_cleaned.csv", index=False)
```

10.

```python
print(df["Url_youtube"].notna().sum())
```

11.

```python
top10_danceable = df.groupby("Artist")["Danceability"].mean().sort_values(ascending=False).head(10)
print(top10_danceable)
```

12.

```python
top10_stream = df.groupby("Artist")["Stream"].sum().sort_values(ascending=False).head(10)
top10_views = df.groupby("Artist")["Views"].sum().sort_values(ascending=False).head(10)

print(top10_stream)
print(top10_views)
```

14.

```python
artists_to_compare = ["J. Cole", "Kendrick Lamar", "Drake"]
df_filtered = df[df["Artist"].isin(artists_to_compare)].drop_duplicates(subset=["Track"])
comparison = df_filtered.groupby("Artist")[["Stream", "Views"]].sum()

most_streams = comparison["Stream"].sort_values(ascending=False).index[0]
most_views = comparison["Views"].sort_values(ascending=False).index[0]

print(most_streams)
print(most_views)
```
